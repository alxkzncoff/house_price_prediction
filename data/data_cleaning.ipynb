```python
def add_district(column):
    
    """ –§–Ω—É–∫—Ü–∏–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞–Ω–Ω—ã—Ö –æ —Å—Ç–∞–Ω—Ü–∏–∏ –º–µ—Ç—Ä–æ, –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≤ –∫–∞–∫–æ–º —Ä–∞–π–æ–Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –∫–≤–∞—Ä—Ç–∏—Ä–∞."""

    districts = {
        '–ê–¥–º–∏—Ä–∞–ª—Ç–µ–π—Å–∫–∏–π':     ['–ë–∞–ª—Ç–∏–π—Å–∫–∞—è', '–§—Ä—É–Ω–∑–µ–Ω—Å–∫–∞—è', '–ó–≤–µ–Ω–∏–≥–æ—Ä–æ–¥—Å–∫–∞—è', '–°–µ–Ω–Ω–∞—è –ø–ª–æ—â–∞–¥—å',
                               '–°–∞–¥–æ–≤–∞—è', '–ü—É—à–∫–∏–Ω—Å–∫–∞—è', '–¢–µ—Ö–Ω–æ–ª–æ–≥–∏—á–µ—Å–∫–∏–π –∏–Ω-—Ç II', '–¢–µ—Ö–Ω–æ–ª–æ–≥–∏—á–µ—Å–∫–∏–π –∏–Ω-—Ç I',
                               '–°–ø–∞—Å—Å–∫–∞—è'],
        '–í–∞c–∏–ª–µ–æ—Å—Ç—Ä–æ–≤—Å–∫–∏–π':   ['–í–∞—Å–∏–ª–µ–æ—Å—Ç—Ä–æ–≤—Å–∫–∞—è', '–ü—Ä–∏–º–æ—Ä—Å–∫–∞—è'],
        '–í—ã–±–æ—Ä–≥—Å–∫–∏–π':         ['–õ–µ—Å–Ω–∞—è', '–£–¥–µ–ª—å–Ω–∞—è', '–û–∑–µ—Ä–∫–∏', '–ü—Ä–æ—Å–ø–µ–∫—Ç –ü—Ä–æ—Å–≤–µ—â–µ–Ω–∏—è', '–í—ã–±–æ—Ä–≥—Å–∫–∞—è',
                               '–ü–∞—Ä–Ω–∞—Å', '–®—É–≤–∞–ª–æ–≤–æ'],
        '–ö–∞–ª–∏–Ω–∏–Ω—Å–∫–∏–π':        ['–ü–ª–æ—â–∞–¥—å –õ–µ–Ω–∏–Ω–∞', '–ê–∫–∞–¥–µ–º–∏—á–µ—Å–∫–∞—è', '–ì—Ä–∞–∂–¥–∞–Ω—Å–∫–∏–π –ø—Ä–æ—Å–ø–µ–∫—Ç', '–ü–ª–æ—â–∞–¥—å –ú—É–∂–µ—Å—Ç–≤–∞',
                               '–†—É—á—å–∏', '–ü–æ–ª–∏—Ç–µ—Ö–Ω–∏—á–µ—Å–∫–∞—è', '–ü–∏—Å–∫–∞—Ä—ë–≤–∫–∞'],
        '–ö–∏—Ä–æ–≤—Å–∫–∏–π':          ['–ü—Ä–æ—Å–ø–µ–∫—Ç –í–µ—Ç–µ—Ä–∞–Ω–æ–≤', '–ù–∞—Ä–≤—Å–∫–∞—è', '–õ–µ–Ω–∏–Ω—Å–∫–∏–π –ø—Ä–æ—Å–ø–µ–∫—Ç', '–ê–≤—Ç–æ–≤–æ', '–ö–∏—Ä–æ–≤—Å–∫–∏–π –∑–∞–≤–æ–¥'],
        '–ö–æ–ª–ø–∏–Ω—Å–∫–∏–π':         [],
        '–ö—Ä–∞—Å–Ω–æ–≥–≤–∞—Ä–¥–µ–π—Å–∫–∏–π':  ['–õ–∞–¥–æ–∂—Å–∫–∞—è', '–ù–æ–≤–æ—á–µ—Ä–∫–∞—Å—Å–∫–∞—è'],
        '–ö—Ä–∞—Å–Ω–æ—Å–µ–ª—å—Å–∫–∏–π':     ['–°–µ—Ä–≥–∏–µ–≤–æ (–í–æ–ª–æ–¥–∞—Ä—Å–∫–∞—è)', '–°–æ—Å–Ω–æ–≤–∞—è –ü–æ–ª—è–Ω–∞', '–ì–æ—Ä–µ–ª–æ–≤–æ', '–õ–∏–≥–æ–≤–æ'],
        '–ö—Ä–æ–Ω—à—Ç–∞–¥—Ç—Å–∫–∏–π':      [],
        '–ö—É—Ä–æ—Ä—Ç–Ω—ã–π':          [],
        '–ú–æ—Å–∫–æ–≤—Å–∫–∏–π':         ['–ó–≤–µ–∑–¥–Ω–∞—è', '–ö—É–ø—á–∏–Ω–æ', '–ú–æ—Å–∫–æ–≤—Å–∫–∏–µ –≤–æ—Ä–æ—Ç–∞', '–ú–æ—Å–∫–æ–≤—Å–∫–∞—è', '–ü–∞—Ä–∫ –ü–æ–±–µ–¥—ã',
                               '–≠–ª–µ–∫—Ç—Ä–æ—Å–∏–ª–∞'],
        '–ù–µ–≤—Å–∫–∏–π':            ['–û–±—É—Ö–æ–≤–æ', '–ï–ª–∏–∑–∞—Ä–æ–≤—Å–∫–∞—è', '–†—ã–±–∞—Ü–∫–æ–µ', '–ü—Ä–æ–ª–µ—Ç–∞—Ä—Å–∫–∞—è', '–£–ª–∏—Ü–∞ –î—ã–±–µ–Ω–∫–æ',
                               '–ü—Ä–æ—Å–ø–µ–∫—Ç –ë–æ–ª—å—à–µ–≤–∏–∫–æ–≤', '–õ–æ–º–æ–Ω–æ—Å–æ–≤—Å–∫–∞—è'],
        '–ü–µ—Ç—Ä–æ–≥—Ä–∞–¥—Å–∫–∏–π':      ['–ö—Ä–µ—Å—Ç–æ–≤—Å–∫–∏–π –æ—Å—Ç—Ä–æ–≤', '–ü–µ—Ç—Ä–æ–≥—Ä–∞–¥—Å–∫–∞—è', '–ì–æ—Ä—å–∫–æ–≤—Å–∫–∞—è', '–ß–∫–∞–ª–æ–≤—Å–∫–∞—è',
                               '–°–ø–æ—Ä—Ç–∏–≤–Ω–∞—è', '–ù–æ–≤–æ–∫—Ä–µ—Å—Ç–æ–≤—Å–∫–∞—è'],
        '–ü–µ—Ç—Ä–æ–¥–≤–æ—Ä—Ü–æ–≤—ã–π':     [],
        '–ü—Ä–∏–º–æ—Ä—Å–∫–∏–π':         ['–ö–æ–º–µ–Ω–¥–∞–Ω—Ç—Å–∫–∏–π –ø—Ä–æ—Å–ø–µ–∫—Ç', '–°—Ç–∞—Ä–∞—è –¥–µ—Ä–µ–≤–Ω—è', '–ü–∏–æ–Ω–µ—Ä—Å–∫–∞—è', '–ß–µ—Ä–Ω–∞—è —Ä–µ—á–∫–∞',
                               '–ë–µ–≥–æ–≤–∞—è', '–û–ª—å–≥–∏–Ω–æ'],
        '–ü—É—à–∫–∏–Ω—Å–∫–∏–π':         [],
        '–§—Ä—É–Ω–∑–µ–Ω—Å–∫–∏–π':        ['–ë—É—Ö–∞—Ä–µ—Å—Ç—Å–∫–∞—è', '–í–æ–ª–∫–æ–≤—Å–∫–∞—è', '–û–±–≤–æ–¥–Ω—ã–π –∫–∞–Ω–∞–ª', '–ú–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω–∞—è'],
        '–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π':        ['–ß–µ—Ä–Ω—ã—à–µ–≤—Å–∫–∞—è', '–ü–ª–æ—â–∞–¥—å –ê. –ù–µ–≤—Å–∫–æ–≥–æ', '–ü–ª–æ—â–∞–¥—å –í–æ—Å—Å—Ç–∞–Ω–∏—è', '–ù–µ–≤—Å–∫–∏–π –ø—Ä–æ—Å–ø–µ–∫—Ç',
                               '–õ–∏–≥–æ–≤—Å–∫–∏–π –ø—Ä–æ—Å–ø–µ–∫—Ç', '–ì–æ—Å—Ç–∏–Ω—ã–π –¥–≤–æ—Ä', '–ê–¥–º–∏—Ä–∞–ª—Ç–µ–π—Å–∫–∞—è', '–î–æ—Å—Ç–æ–µ–≤—Å–∫–∞—è', '–ú–∞—è–∫–æ–≤—Å–∫–∞—è',
                               '–í–ª–∞–¥–∏–º–∏—Ä—Å–∫–∞—è'],
        '–í—Å–µ–≤–æ–ª–æ–∂—Å–∫–∏–π':       ['–î–µ–≤—è—Ç–∫–∏–Ω–æ', '–®–æ—Å—Å–µ–π–Ω–∞—è']
    }

    for key, value in districts.items():
        if column in value:
            return key
```

```python
def data_prep(df):
    
    """ –§—É–Ω–∫—Ü–∏—è –ø—Ä–µ–¥. –æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö. """

    temp_df = df.copy()  # –ö–æ–ø–∏—Ä—É–µ–º dataframe, —á—Ç–æ–±—ã –Ω–µ–±—ã–ª–æ –æ—à–∏–±–∫–∏ SettingWithCopyWarning
    
    temp_df.drop_duplicates(inplace=True) # —É–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–µ–π
        
#   price (target) - —É–±–∏—Ä–∞–µ–º –ª–∏—à–Ω—é—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é (P). –ú–µ–Ω—è–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–∞ float64.
    temp_df['price'] = temp_df['price'].apply(lambda x: float(x.replace(' ','')[:-1]))
    
#   price_per_m - –∏–º–µ–µ—Ç —Ç–æ–∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —á—Ç–æ –∏ —Ç–∞—Ä–≥–µ—Ç.
    temp_df.drop(columns=['price_per_m'], inplace=True)

#   flat_type - –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ø—Ä–∏–Ω–∞–∫–∞ rooms –¥–æ–±–∞–≤–ª—è–µ–º –¥–≤–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ "–°—Ç—É–¥–∏—è"
#               –∏ "–°–≤–æ–±. –ª–∞–Ω–∏—Ä–æ–≤–∫–∞"
    temp_df['flat_type'] = temp_df['rooms'].apply(lambda x: '–°—Ç—É–¥–∏—è' if x == '–°—Ç—É–¥–∏—è' else
                                                  ('–°–≤–æ–±. –ø–ª–∞–Ω–∏—Ä–æ–≤–∫–∞' if x == '–°–≤–æ–±. –ø–ª–∞–Ω–∏—Ä–æ–≤–∫–∞' else '–ö–≤–∞—Ä—Ç–∏—Ä–∞'))
    
#   rooms - –º–µ–Ω—è–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏—é "–°—Ç—É–¥–∏—è" –Ω–∞ "1". –≠—Ç–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è —É–∂–µ –µ—Å—Ç—å –≤ –ø—Ä–∏–∑–Ω–∞–∫–µ flat_type.
#   –ß–∏—Å—Ç–∏–º –¥–∞–Ω–Ω—ã–µ –æ—Ç –º—É—Å–æ—Ä–∞.
    temp_df['rooms'] = temp_df['rooms'].apply(lambda x: '9' if x=='&gt; 9' else x)
    temp_df['rooms'] = temp_df['rooms'].apply(lambda x: '1' if x=='–°—Ç—É–¥–∏—è' else x)
    
#   foolrs - –¥–µ–ª–∞–µ–º –∏–∑ –æ–¥–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –¥–≤–∞. floors - —ç—Ç–∞–∂ –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä–∞.
#   total_floors - –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç—Ç–∞–∂–µ–π –≤ –¥–æ–º–µ.
    temp_df['total_floors'] = None
    temp_df['total_floors'] = temp_df['floors'].apply(lambda x: x.split('/')[1])
    temp_df['floors'] = temp_df['floors'].apply(lambda x: x.split('/')[0])
    
#   square - —É–±–∏—Ä–∞–µ–º –ª–∏—à–Ω—é—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é (–º2). –ú–µ–Ω—è–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–∞ float64.
    temp_df['square'] = temp_df['square'].apply(lambda x: float(x.split(' ')[0]))
    
#   kitchen_square - —É–±–∏—Ä–∞–µ–º –ª–∏—à–Ω—é—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é (–º2). –ú–µ–Ω—è–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–∞ float64.
    temp_df['kitchen_square'] = temp_df['kitchen_square'].apply(lambda x: None if type(x)==float else float(x[:-3]))
    
#   live_square - —É–±–∏—Ä–∞–µ–º –ª–∏—à–Ω—é—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é (–º2). –ú–µ–Ω—è–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–∞ float64.
    temp_df['live_square'] = temp_df['live_square'].apply(lambda x: None if type(x)==float else float(x[:-3]))
    
#   underground - –ß–∏—Å—Ç–∏–º –¥–∞–Ω–Ω—ã–µ, —É–¥–∞–ª—è–µ–º –ø–æ–≤—Ç–æ—Ä—ã –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ—á–∫–µ. –î–µ–ª–∞–µ–º –∏–∑ –æ–¥–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –¥–≤–∞.
#   underground - —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–ª—è –±–ª–∏–∂–∞–π—à–µ–π —Å—Ç–∞–Ω—Ü–∏–∏ –º–µ—Ç—Ä–æ. metro_station - –Ω–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–∞–Ω—Ü–∏–∏. 
#   –ü—Ä–∏–≤–æ–¥–∏–º –≤—Å–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –∫ –º–µ—Ç—Ä–∞–º.
    temp_df['underground'] = temp_df['underground'].apply(lambda x: x.split(',')[0] if type(x)==str else x)
    
    temp_df['metro_station'] = None
    temp_df['metro_station'] = temp_df['underground'].apply(lambda x: ' '.join(x.split()[:3]) if type(x) == str and len(x.split()) > 4 else
                                                                     (' '.join(x.split()[:2]) if type(x) == str and len(x.split()) == 4 else 
                                                                     (x.split()[0] if type(x) == str and len(x.split()) == 3 else x)))
    
    temp_df['underground'] = temp_df['underground'].apply(lambda x: float(x.split()[-2]) if type(x) == str and x.split()[-1] == '–º' else
                                                                   (float(x.split()[-2])*1000 if type(x) == str and x.split()[-1] == '–∫–º' else x))

#   –æ–±—ä—è–≤–ª–µ–Ω–∏—è —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –¥–æ —Å—Ç–∞–Ω—Ü–∏–∏ –º–µ—Ç—Ä–æ 0 —Å–æ–¥–µ—Ä–∂–∞—Ç –Ω–µ —Ç–æ—á–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏ –∏—Ö –º–∞–ª–æ. –ü—Ä–æ—Å—Ç–æ —É–¥–∞–ª—è–µ–º.
    temp_df.drop(index=(temp_df[temp_df['underground']==0].index), inplace=True)
    
#   –æ–±—ä—è–≤–ª–µ–Ω–∏ –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ –ª–µ–Ω. –æ–±–ª–∞—Å—Ç–∏ –∏–º–µ—é—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ –º–µ—Ç—Ä–æ –±–æ–ª–µ–µ 5–∫–º. –î–µ–ª–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ –∫–∞—Ç–µ–≥–æ–≥—Ä–∏–∞–ª—å–Ω—ã–º.    
    temp_df['underground'] = temp_df['underground'].apply(lambda x: '0 - 1000' if x < 1001 else 
                                                          ('1000 - 2000' if x > 1000 and x < 2001 else
                                                          ('2000 - 3000' if x > 2000 and x < 3001 else
                                                          ('3000 - 4000' if x > 3000 and x < 4001 else
                                                          ('4000 - 5000' if x > 4000 and x < 5001 else x)))))
    
#   –≤—Å–µ –ø—Ä–æ–ø—É—Å–∫–∏ –∑–∞–ø–æ–ª–Ω–∏–º "> 5000", —Ç–∞–∫ –∫–∞–∫ –æ–±—ä—è–≤–ª–µ–Ω–∏—è –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ –ª–µ–Ω. –æ–±–ª–∞—Å—Ç–∏.
    temp_df['underground'].fillna('> 5000', inplace=True)
    
#   –æ–ø—Ä–µ–¥–µ–ª—è–µ–º —Ä–∞–π–æ–Ω –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ district –∏ mmetro_station. 
#   –æ–ø—Ä–µ–¥–µ–ª—è–µ–º —Ä–∞–π–æ–Ω –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ district.
    temp_df['district_by_district_feature'] = 0
    temp_df['district_by_district_feature'] = temp_df['district'].apply(lambda x: x.split(',')[1] if len(x.split(','))>1 else x)
    temp_df['district_by_district_feature'] = temp_df['district_by_district_feature'].apply(lambda x: x.split()[0] if len(x.split())>1 else x)
    
#   –æ–ø—Ä–µ–¥–µ–ª—è–µ–º —Ä–∞–π–æ–Ω –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ metro_station.
    temp_df['district'] = temp_df['metro_station'].apply(add_district)
    
#   –∑–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ district_by_district_feature.
    temp_df['district'].fillna(value=temp_df['district_by_district_feature'], inplace=True)
    
#   –æ—Å—Ç–∞–≤–ª—è–µ–º –æ–¥–∏–Ω –ø—Ä–∏–∑–Ω–∞–∫ district.
    temp_df.drop(columns='district_by_district_feature', inplace=True)
    
#   –æ–±—ä—è–≤–ª–µ–Ω–∏—è –±–µ–∑ —Å—Ç–∞–Ω—Ü–∏–∏ –º–µ—Ç—Ä–æ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –∑–∞ –ø—Ä–µ–¥–µ–ª–∞–º–∏ –≥–æ—Ä–æ–¥–∞.
    temp_df['metro_station'].fillna('–õ–µ–Ω. –æ–±–ª–∞—Å—Ç—å', inplace=True)
    
#   —É–¥–∞–ª—è–µ–º catalod_id.
    temp_df.drop(columns=['catalog_id'], inplace=True)

    return temp_df
```
