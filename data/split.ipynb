# TRAIN TEST SPLIT

**–°—Ä–∞–∑—É –∂–µ —Ä–∞–∑–¥–µ–ª—é –¥–∞–Ω–Ω—ã–µ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ. –°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ 70/30. –û–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –∏—Ö –±—É–¥—É –æ—Ç–¥–µ–ª—å–Ω–æ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.** 

```python
train, test = train_test_split(data, test_size=0.3, shuffle=True)
```

```python
train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
```

**–°—Ä–∞–∑—É –∂–µ —É–¥–∞–ª—é html —Ç–µ–≥–∏ `<span>`, `</span>`.**

```python
def remove_span(df):
    
    """–§—É–Ω–∫—Ü–∏—è —É–¥–∞–ª—è–µ—Ç html —Ç–µ–≥–∏ –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞."""
    
    temp_df = df.copy()  # –ö–æ–ø–∏—Ä—É–µ–º dataframe, —á—Ç–æ–±—ã –Ω–µ–±—ã–ª–æ –æ—à–∏–±–∫–∏ SettingWithCopyWarning
    
    for column in temp_df.columns[2:-13]:
        temp_df[column] = temp_df[column].apply(lambda x: x.split('<span>')[1].split('</span>')[0])
        
    return temp_df
```

```python
train = remove_span(train)
test = remove_span(test)
```
