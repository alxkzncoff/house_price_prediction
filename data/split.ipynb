При повторном сборе данных, в одной строчке данные дублируют название фичей. Добавил дополнительный этап обработки.

# TRAIN TEST SPLIT

**Сразу же разделю данные на тренировочные и тестовые. Соотношение 70/30. Обрабатывать их буду отдельно друг от друга.** 

```python
train, test = train_test_split(data, test_size=0.3, shuffle=True)
```

```python
train.head(5)
```

```python
test.head(5)
```

```python
train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
```

**Сразу же удалю html теги `<span>`, `</span>`.**

```python
def remove_span(df):
    
    """Функция удаляет html теги из датасета."""
    
    temp_df = df.copy()  # Копируем dataframe, чтобы небыло ошибки SettingWithCopyWarning
    
    for column in temp_df.columns[2:-13]:
        temp_df[column] = temp_df[column].apply(lambda x: x.split('<span>')[1].split('</span>')[0])
        
    return temp_df
```

```python
train = remove_span(train)
test = remove_span(test)
```
